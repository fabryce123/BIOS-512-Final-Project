For my project I will be analyzing my favorite basketball player, LeBron James, statistics throughout his 2012-2013 season, often considered his best. All of my data is coming from https://www.basketball-reference.com/. In my analysis, I am comparing LeBron's statistics in games he won versus games he lost. For convenience, I have split the dataset for Lebron's statists when he losses and wins.

```r
library("dplyr")
library("tidyverse")
library("ggplot2")
library("patchwork")
lebron_master<-read.csv("lebron all games.csv")
lebron_wins<-read.csv("LBJ.csv")
lebron_losses<-read.csv("LBJ losses.csv")
```

Here is the head of the table for wins and losses.

```r
head(lebron_wins)
head(lebron_losses)
```

One of the key attributes of a great NBA player and winning games is the ability to shoot the basketball with good efficiency. A player that makes baskets at a high percentage as their shot attempts increase is an efficient shooter. Below I am looking at a scatterplot of LeBron's three-pointer shot attempts and three-pointer shooting percentage in wins and losses. It is clear that in wins there was a somewhat positive relationship between LeBron's three-pointer shot attempts and three-pointer shooting percentage, while in losses there was a somewhat negative relationship between LeBron's three-pointer shot attempts and three-pointer shooting percentage.

```r
ggplot(lebron_wins, aes(x=X3PA, y=`X3P.`)) + geom_point() +labs(y= "Three Pointer Shot Attempts", x = "Three Pointer Shooting Percentage", title="Three pointer Attempts vs Percentage in Wins")
ggplot(lebron_losses, aes(x=X3PA, y=`X3P.`)) + geom_point()+labs(y= "Three Pointer Shot Attempts", x = "Three Pointer Shooting Percentage", title="Three pointer Attempts vs Percentage in Losses")
```

Next, I am curious about the total points LeBron scored in wins compared to losses. I think a boxplot is a good visualization to use for this since I'd like to compare the distribution of the continuous values between the categories in my data. From the plot below, it is obvious that the median points scored in wins are higher than the median points scored in losses. This makes sense because LeBron is an integral part of his team and if he is not scoring many points, his team is more likely to lose.

```r
ggplot(lebron_master, aes(x=Status, y=PTS, fill=Status)) + geom_boxplot() +labs(y= "Points Scored", x = "Win or Loss", title="Points scored in Wins vs Losses")
```
